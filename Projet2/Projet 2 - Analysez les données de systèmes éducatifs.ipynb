Cette étape sert à vérifier si les colonnes identifiées comme **clés potentielles** comportent réellement des valeurs uniques (i.e. aucun doublon) afin de permettre de faire des `merge`.  On ne touche pas encore aux valeurs, on vérifie juste la cohérence théorique des dimensions.

On a utilisé deux méthodes principales :

---

#### A. a. Affichage des dimensions des DataFrames & calcul du nombre de valeurs uniques

##### A. a. i. La colonne `SeriesCode` du df_series vs la colonne `Indicator Code` du df_data

- `df_series` possède une clé exploitable `SeriesCode` qui **ne comporte aucun doublon**.  
  - Exemple : sur **3665 lignes au total**, la colonne `SeriesCode` contient **3665 valeurs uniques**.
  
##### A. a. ii. Merge des df_data, df_country_new grâce à la colonne `CountryCode`

> La colonne `Indicator Code` de `df_data` contient le **même nombre de valeurs uniques (3665)** que la colonne `SeriesCode` de `df_series`.

---

#### A. b. Utilisation des méthodes `.duplicated()` et `.any()`

- `.duplicated()` → retourne un **booléen pour chaque ligne** indiquant si la valeur est dupliquée.  
- `.any()` → retourne un **booléen global** :
  - `True` : la colonne contient des valeurs dupliquées (non uniques)
  - `False` : la colonne ne contient que des valeurs uniques

## III. 1. Filtrage des cellules à fort taux de remplissage

### A. Filtrage des colonnes par le taux de valeurs de manquantes

```python
df_data_final.shape
```

```python
pd.set_option('display.max_rows', 150)
df_NaN = pd.DataFrame(df_data_final.isna().mean()*100).reset_index().sort_values(by=0, ascending=False)
display(df_NaN) 
# on crée un DataFrame booléen (True si NaN, False sinon).

# .mean() permet de calculer la moyenne des True (où True = 1)

# ce qui nous donne la proportion de valeurs manquantes par colonne.
```

```python
import plotly.express as px

# DÉFINITION DU SEUIL

# Identifier les colonnes années (triées numériquement)
year_cols = sorted(
    [col for col in df_data_final.columns if col.isdigit()],
    key=int
)

# Calcul du pourcentage de valeurs manquantes par année
df_NaN_years = df_data_final[year_cols].isna().mean() * 100
df_NaN_years = df_NaN_years.reset_index()
df_NaN_years.columns = ['Year', 'Missing_pct']

# Seuil
threshold = 88

# Création du barplot interactif
fig = px.bar(
    df_NaN_years,
    x='Year',
    y='Missing_pct',
    text=df_NaN_years['Missing_pct'].round(1),  # labels sur les barres
    labels={'Missing_pct': 'Pourcentage de valeurs manquantes (%)', 'Year': 'Année'},
    color='Missing_pct',
    color_continuous_scale='Blues'
)

# Ligne du seuil
fig.add_hline(
    y=threshold,
    line_dash="dash",
    line_color="red",
    annotation_text=f"Seuil {threshold}%",
    annotation_position="top right"
)

# Mise en forme
fig.update_layout(
    title="Taux de valeurs manquantes sur les colonnes Années",
    xaxis_tickangle=-90,
    yaxis=dict(range=[0, 100]),
    coloraxis_showscale=False
)

# Affichage interactif
fig.show()

# Export en HTML
fig.write_html("missing_values_barplot.html", include_plotlyjs='cdn', full_html=True)

print("Graphique exporté dans missing_values_barplot.html")
```

Cette visualisation en barres, centrée sur les colonnes correspondant aux années, permet d’identifier un seuil de filtrage pertinent distinguant les années les mieux renseignées de celles comportant trop de valeurs manquantes.

Avec un **seuil fixé à 88 %**, on observe une **série d’années consécutives (de 1999 à 2015) présentant moins de 88 % de valeurs manquantes**, c’est-à-dire les années contenant le plus d’informations exploitables pour notre étude.

Ainsi, on peut choisir de sélectionner non seulement ces colonnes annuelles, mais également toutes les autres variables dont le taux de valeurs manquantes est inférieur à 88 %, afin d’affiner l’analyse et de se concentrer sur les données les plus complètes.

```python
# Selection d'une liste de colonnes à garder contenant le moins de valeurs manquantes
col_to_keep = list(df_NaN[df_NaN[0]< threshold]['index'].unique()) 

df_filtered = df_data_final[col_to_keep].copy()
# Nouveau dataframe filtré

print(f"✅ Nouveau dataframe filtré : {df_filtered.shape[1]} colonnes conservées (contre {df_data_final.shape[1]} avant filtrage)")
```

```python
df_filtered.head(5)
```

```python
# Réorganisation de l'ordre des colonnes

# 1️⃣ Colonnes descriptives à garder en premier
colonnes_avant = [
    'Country Name', 'CountryCode', 'SeriesCode',
    'Indicator Name_x', 'Topic', 'Long definition'
]

# 2️⃣ Identifier les colonnes années (en les triant numériquement)
year_cols = sorted(
    [col for col in df_filtered.columns if col.isdigit()],
    key=int
)

# 3️⃣ Identifier les autres colonnes (ni descriptives principales, ni années)
autres_colonnes = [
    col for col in df_filtered.columns
    if col not in colonnes_avant + year_cols
]

# 4️⃣ Réorganisation finale
df_filtered = df_filtered[colonnes_avant + year_cols + autres_colonnes]

# 5️⃣ Vérification
df_filtered.head(5)
```

Par rapport à notre problématique, trois dimensions sont essentielles à l’accès et à l’investissement dans l’éducation en ligne:

**1. Accessibilité à la connexion Internet**

Avant toute chose, si l'on souhaite identifier les pays (ou territoires) dans lesquels la formation en ligne  de niveau secondaire et tertiaire pourrait recevoir un bon potentiel de clients intéressés, il faut s'assurer au préalable que le taux d'accès à internet est assez élevé. 

Une large couverture de l’indicateur `Internet users (per 100 people)` permet ainsi d’évaluer le niveau de connectivité de la population et donc la capacité réelle à accéder à des contenus éducatifs en ligne.

**2. Niveau de richesse et capacité à investir**

Pour savoir si les étudiants d’un pays ont les moyens financiers d’investir dans leurs études — notamment en ligne —, il faut vérifier deux éléments :

- **Pouvoir d’achat réel**  
  → Combien de biens et services (dont l’éducation) un étudiant peut se payer localement.

- **Revenus des résidents**  
  → Les revenus réellement disponibles pour les habitants du pays.

Le **GNI per capita, PPP (current international $)** est le plus pertinent ici, car il représente **le revenu moyen par habitant**, **corrigé du coût de la vie**.

En revanche, le **GDP per capita (PPP)** mesure **la richesse produite sur le territoire**,  qui **n’est pas forcément captée par les résidents**.  
Dans certains pays, une part importante de la richesse est créée par des entreprises étrangères, sans nécessairement bénéficier à la population locale.

Les indicateurs économiques tels que `GNI per capita, PPP (current international $)`, `Population, ages 15-64, total`, `Population, ages 15-24, total` et `Population growth (annual %)` offrent une vision complémentaire du pouvoir d’achat et de la taille de la population active.

Combinés, ils permettent d’estimer la population potentiellement capable d’investir dans des formations en ligne.

**3. Niveau global d’éducation**

Enfin, l'indicateur `Enrolment in secondary general, both sexes (number)` relatif à la scolarisation permet d’évaluer le niveau d’éducation atteint et donc la prédisposition à suivre des cours en ligne plus avancés.

#### C. ii. Agrégation des pays les mieux classés sur l’ensemble des indicateurs

Cette section regroupe les pays qui apparaissent dans le top 5 pour chacun des indicateurs précédemment analysés, afin d’identifier ceux qui sont globalement les plus favorables pour l’implantation de la start-up EdTech.

```python
# Exemple de poids
weights = {
    'Internet': 0.4,                # Très important
    'GNI': 0.3,                     # Assez important
    'Secondary Enrollment': 0.2,    # Moyen
    'Population 15-24': 0.1         # Non essentiel
}

# Tes listes top 5 par indicateur
top_5 = {
    'Internet': top_countries_Internet,
    'GNI': top_countries_GNI,
    'Secondary Enrollment': top_countries_education,
    'Population 15-24': top_countries_age
}

top_5
```

```python
from collections import defaultdict

scores = defaultdict(float)

for indicator, countries in top_5.items():
    w = weights[indicator]
    for rank, country in enumerate(countries):
        # 5 - rank donne 5 pour le premier, 1 pour le dernier
        scores[country] += (len(countries) - rank) * w
```

```python
weights = {
    "Individuals using the Internet (% of population)": 0.4,
    "GNI per capita, PPP (current international $)": 0.3,
    "Enrolment in secondary general, both sexes (number)": 0.2,
    "Population ages 15-24, total": 0.1
}

df_weighted = df_long[df_long['Indicator Name_x'].isin(weights.keys())].copy()

df_weighted['Value_norm'] = df_weighted.groupby('Indicator Name_x')['Value'] \
                                       .transform(lambda x: (x - x.min()) / (x.max() - x.min()))

df_weighted['Weight'] = df_weighted['Indicator Name_x'].map(weights)
df_weighted['Weighted_value'] = df_weighted['Value_norm'] * df_weighted['Weight']

score_by_country_year = df_weighted.groupby(['CountryCode', 'Year'])['Weighted_value'] \
                                   .mean() \
                                   .reset_index(name='Score')

selected_countries = ['ISL','NOR','LUX','MAC','USA','KWT','IDN','BRN',
                      'DNK','VNM','SGP','IND','BMU','RUS','CHN','TUR','PAK']

score_plot = score_by_country_year[score_by_country_year['CountryCode'].isin(selected_countries)]


import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

plt.figure(figsize=(12, 6))

# palette avec autant de couleurs que de pays
colors = cm.get_cmap('tab20', len(selected_countries))

for i, country in enumerate(selected_countries):
    df_country = score_plot[score_plot['CountryCode'] == country]
    if not df_country.empty:
        plt.plot(df_country['Year'], df_country['Score'], 
                 marker='o', 
                 label=country,
                 color=colors(i))

plt.title("Score global pondéré par pays (tous indicateurs combinés)")
plt.xlabel("Année")
plt.ylabel("Score pondéré (normalisé)")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
