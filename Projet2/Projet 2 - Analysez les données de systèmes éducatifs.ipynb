- `.duplicated()` → retourne un **booléen pour chaque ligne** indiquant si la valeur est dupliquée.

### A. Filtrage des colonnes par le taux de valeurs de manquantes

```python
df_data_final.shape
```

```python
pd.set_option('display.max_rows', 150)
df_NaN = pd.DataFrame(df_data_final.isna().mean()*100).reset_index().sort_values(by=0, ascending=False)
display(df_NaN) 
# on crée un DataFrame booléen (True si NaN, False sinon).

# .mean() permet de calculer la moyenne des True (où True = 1)

# ce qui nous donne la proportion de valeurs manquantes par colonne.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# DÉFINITION DU SEUIL

# Identifier les colonnes années (triées numériquement)
year_cols = sorted(
    [col for col in df_data_final.columns if col.isdigit()],
    key=int
)

# Calcul du pourcentage de valeurs manquantes par année
df_NaN_years = df_data_final[year_cols].isna().mean() * 100  # % de NaN
df_NaN_years = df_NaN_years.reset_index()
df_NaN_years.columns = ['Year', 'Missing_pct']

# Tracé du barplot 
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=df_NaN_years, x="Year", y="Missing_pct", color="skyblue")

# Ligne horizontale du seuil
threshold = 88  # Exemple : seuil de 88 % de valeurs manquantes
plt.axhline(y=threshold, color='red', linestyle='--', linewidth=2, label=f'Seuil {threshold}%')

# Mise en forme
plt.xticks(rotation=45, ha='right')
plt.title("Taux de valeurs manquantes sur les colonnes Années", fontsize=14)
plt.xlabel("Année")
plt.ylabel("Pourcentage de valeurs manquantes (%)")

# Ajout des labels sur chaque barre
for container in ax.containers:
    ax.bar_label(
        container,
        fmt='%.1f',
        padding=3,
        fontsize=8,
        rotation=90
    )

# Légende
plt.legend()

plt.tight_layout()
plt.show()
```

Cette visualisation en barres, centrée sur les colonnes correspondant aux années, permet d’identifier un seuil de filtrage pertinent distinguant les années les mieux renseignées de celles comportant trop de valeurs manquantes.

Avec un **seuil fixé à 88 %**, on observe une **série d’années consécutives (de 1999 à 2015) présentant moins de 88 % de valeurs manquantes**, c’est-à-dire les années contenant le plus d’informations exploitables pour notre étude.

Ainsi, on peut choisir de sélectionner non seulement ces colonnes annuelles, mais également toutes les autres variables dont le taux de valeurs manquantes est inférieur à 88 %, afin d’affiner l’analyse et de se concentrer sur les données les plus complètes.

```python
# Selection d'une liste de colonnes à garder contenant le moins de valeurs manquantes
col_to_keep = list(df_NaN[df_NaN[0]< threshold]['index'].unique()) 

df_filtered = df_data_final[col_to_keep].copy()
# Nouveau dataframe filtré

print(f"✅ Nouveau dataframe filtré : {df_filtered.shape[1]} colonnes conservées (contre {df_data_final.shape[1]} avant filtrage)")
```

```python
df_filtered.head(5)
```

```python
# Réorganisation de l'ordre des colonnes

# 1️⃣ Colonnes descriptives à garder en premier
colonnes_avant = [
    'Country Name', 'CountryCode', 'SeriesCode',
    'Indicator Name_x', 'Topic', 'Long definition'
]

# 2️⃣ Identifier les colonnes années (en les triant numériquement)
year_cols = sorted(
    [col for col in df_filtered.columns if col.isdigit()],
    key=int
)

# 3️⃣ Identifier les autres colonnes (ni descriptives principales, ni années)
autres_colonnes = [
    col for col in df_filtered.columns
    if col not in colonnes_avant + year_cols
]

# 4️⃣ Réorganisation finale
df_filtered = df_filtered[colonnes_avant + year_cols + autres_colonnes]

# 5️⃣ Vérification
df_filtered.head(5)
```

### A. Choix de l'indicateur GNI per capita, PPP (current international $)

Pour savoir si les étudiants d’un pays ont les moyens financiers d’investir dans leurs études — notamment en ligne —, il faut vérifier deux éléments :

- **Pouvoir d’achat réel**  
  → Combien de biens et services (dont l’éducation) un étudiant peut se payer localement.

- **Revenus des résidents**  
  → Les revenus réellement disponibles pour les habitants du pays.

Le **GNI per capita, PPP (current international $)** est le plus pertinent ici,  
car il représente **le revenu moyen par habitant**, **corrigé du coût de la vie**.

En revanche, le **GDP per capita (PPP)** mesure **la richesse produite sur le territoire**,  
qui **n’est pas forcément captée par les résidents**.  
Dans certains pays, une part importante de la richesse est créée par des entreprises étrangères,  
sans nécessairement bénéficier à la population locale.
